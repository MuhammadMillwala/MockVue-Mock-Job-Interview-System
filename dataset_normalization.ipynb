```python
import pandas as pd
import string
import re
import numpy as np
```

```python
df = pd.read_csv('data/Dataset.csv')
```

```python
df = df.fillna(value=np.nan)
```

```python
df
```

```python
df =  df.drop('Unnamed: 2', axis=1)
df =  df.drop('Unnamed: 3', axis=1)
```

```python
df.fillna(value='None', inplace=True)
```

```python
df = df.dropna()
```

```python
df.head(20)
```

```python
df['Answers'] = df['Answers'].fillna('').astype(str)
df['Answers'] = df['Answers'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df['Answers'] = df['Answers'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df['Answers'] = df['Answers'].apply(lambda x: x.lower())
df['Answers'] = df['Answers'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())

df['Questions'] = df['Questions'].fillna('').astype(str)
df['Questions'] = df['Questions'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df['Questions'] = df['Questions'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
df['Questions'] = df['Questions'].apply(lambda x: x.lower())
df['Questions'] = df['Questions'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
```

```python
df.to_csv('data/normalized_dataset.csv', index=False)
```
